### Importación de datos



```python
import pandas as pd

url = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_1%20.csv"
url2 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_2.csv"
url3 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_3.csv"
url4 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_4.csv"

tienda1 = pd.read_csv(url)
tienda2 = pd.read_csv(url2)
tienda3 = pd.read_csv(url3)
tienda4 = pd.read_csv(url4)

tienda1.head()
```

```python
#Dependencias
%pip install matplotlib
import matplotlib.pyplot as plt
import numpy as np
```

#1. Análisis de facturación



```python
# Datos
categorias = ['Tienda 1', 'Tienda 2', 'Tienda 3', 'Tienda 4']
valores = [sum(tienda1['Precio']), sum(tienda2['Precio']), sum(tienda3['Precio']), sum(tienda4['Precio'])]

# Crear gráfico de barras
plt.bar(categorias, valores)

# Agregar títulos y etiquetas
plt.xlabel('Sucursales')
plt.ylabel('Ganancia total')
plt.title('Totales de ganacias')

# Mostrar el gráfico
plt.show()
```

# 2. Ventas por categoría

```python
# Categorías totales
categorias = list(set(tienda1['Categoría del Producto']))
categoriasTotalesTienda1 = dict()
categoriasTotalesTienda2 = dict()
categoriasTotalesTienda3 = dict()
categoriasTotalesTienda4 = dict()

# Guardar datos de cantidades de categorias
for categoria in categorias:
    categoriasTotalesTienda1[categoria] = list(tienda1['Categoría del Producto']).count(categoria)
    categoriasTotalesTienda2[categoria] = list(tienda2['Categoría del Producto']).count(categoria)
    categoriasTotalesTienda3[categoria] = list(tienda3['Categoría del Producto']).count(categoria)
    categoriasTotalesTienda4[categoria] = list(tienda4['Categoría del Producto']).count(categoria)

dataGraphics = [categoriasTotalesTienda1.values(), categoriasTotalesTienda2.values(), categoriasTotalesTienda3.values(), categoriasTotalesTienda4.values()]

for i, data in enumerate(dataGraphics):
    # Crear gráfico de pastel
    plt.pie(data, labels=categorias, autopct='%1.1f%%', startangle=90)

    # Agregar título
    plt.title(f'Distribución de Ventas por Categoría en Tienda {i + 1}')

    # Mostrar el gráfico
    plt.show()
```

# 3. Calificación promedio de la tienda


```python
tiendas = ['Tienda 1', 'Tienda 2', 'Tienda 3', 'Tienda 4']
# Extraer datos
valores = [sum(tienda1['Calificación']) / len(tienda1['Calificación']), sum(tienda2['Calificación']) / len(tienda2['Calificación']), sum(tienda3['Calificación']) / len(tienda3['Calificación']), sum(tienda4['Calificación']) / len(tienda4['Calificación'])]

# Crear gráfico de líneas
plt.plot(tiendas, valores, marker='o', linestyle='-', color='b')

# Agregar títulos y etiquetas
plt.xlabel('Sucursales')
plt.ylabel('Promedio de Satisfacción')
plt.title('Tendencia de Satisfacción de Clientes')

# Mostrar el gráfico
plt.show()

# Crear gráfico de barras
plt.bar(tiendas, valores)

# Agregar títulos y etiquetas
plt.xlabel('Sucursales')
plt.ylabel('Tendencia de Satisfacción de Clientes')
plt.title('Promedio de Satisfacción')

# Ajustar el rango del eje Y y los intervalos
plt.ylim(3.88, 4.05)
plt.yticks(np.arange(3.88, 4.05, 0.01))  

# Mostrar el gráfico
plt.show()
```

# 4. Productos más y menos vendidos

```python
# Categorías totales
productos = list(set(tienda1['Producto']))
productosTotalesTienda1 = dict()
productosTotalesTienda2 = dict()
productosTotalesTienda3 = dict()
productosTotalesTienda4 = dict()

# Guardar datos de cantidades de categorias
for producto in productos:
    productosTotalesTienda1[producto] = list(tienda1['Producto']).count(producto)
    productosTotalesTienda2[producto] = list(tienda2['Producto']).count(producto)
    productosTotalesTienda3[producto] = list(tienda3['Producto']).count(producto)
    productosTotalesTienda4[producto] = list(tienda4['Producto']).count(producto)

dataGraphics = [productosTotalesTienda1, productosTotalesTienda2, productosTotalesTienda3, productosTotalesTienda4]

for i, data in enumerate(dataGraphics):
    # Ordenar los productos de mayor a menor vendidos
    productosOrdenados = sorted(data.items(), key=lambda x: x[1], reverse=True)
    productos, valores = zip(*productosOrdenados)

    # Crear el histograma
    plt.figure(figsize=(12, 6))  
    plt.bar(productos, valores, color='royalblue')

    # Etiquetas y título
    plt.xlabel('Productos')
    plt.ylabel('Cantidad Vendida')
    plt.title(f'Histograma de Productos más Vendidos en tienda {i + 1}')
    
    # Rotar etiquetas para mayor legibilidad
    plt.xticks(rotation=90)

    # Mostrar el gráfico
    plt.show()

```

# 5. Envío promedio por tienda

```python
tiendas = ['Tienda 1', 'Tienda 2', 'Tienda 3', 'Tienda 4']
# Extraer datos
valores = [sum(tienda1['Costo de envío']) / len(tienda1['Costo de envío']), sum(tienda2['Costo de envío']) / len(tienda2['Costo de envío']), sum(tienda3['Costo de envío']) / len(tienda3['Costo de envío']), sum(tienda4['Costo de envío']) / len(tienda4['Costo de envío'])]

# Crear gráfico de barras
plt.bar(tiendas, valores)

# Agregar títulos y etiquetas
plt.xlabel('Sucursales')
plt.ylabel('Costo de envío promedio')
plt.title('Promedio de envíos')

# Mostrar el gráfico
plt.show()
```

# 6. Análisis del desempeño geográfico

```python
tienda = [tienda1, tienda2, tienda3, tienda4]

for i in range(len(tienda)):
    latitudes = list(tienda[i]['lat'])
    longitudes = list(tienda[i]['lon'])
    costo = list(tienda[i]['Precio'])

    plt.figure(figsize=(10, 6))
    scatter = plt.scatter(longitudes, latitudes, c=costo, cmap='coolwarm', s=100, alpha=0.7)

    plt.colorbar(scatter, label='Ganancia')
    plt.title(f'Mapa de dispersión de ventas por ubicación tienda {i + 1}')
    plt.xlabel('Longitud')
    plt.ylabel('Latitud')
    plt.grid(True)
    plt.show()


```

## Análisis de resultados

### Ingresos totales por tienda




Como se observa, la Tienda 1 lidera en ingresos totales, superando por un pequeño margen a la Tienda 2 y Tienda 3. La Tienda 4 se destaca por su bajo desempeño, lo que podría indicar una menor demanda o problemas operativos.

---

### Categorías más y menos vendidas






Tienda 1 y Tienda 3 tienen un foco fuerte en Muebles y Electrónicos, lo que sugiere un portafolio orientado a bienes de mayor valor.

Tienda 2 presenta una distribución más equilibrada, aunque con menor peso en categorías dominantes.

Tienda 4 tiene la mayor proporción en Juguetes y Artículos para el hogar, lo que puede indicar un enfoque en productos más económicos o de alta rotación.

---

### Calificaciones promedio por tienda



Tienda 3 se destaca como la mejor evaluada por los clientes, superando ligeramente a las demás.

Tienda 1, en cambio, tiene la calificación más baja, lo cual puede reflejar problemas en servicio, producto o experiencia de compra.

La satisfacción del cliente es un factor clave en la rentabilidad a largo plazo. Una tienda con mala calificación podría estar enfrentando más devoluciones, menos recomendaciones o pérdida de clientes recurrentes.

---

### Productos más y menos vendidos






- Tienda 1: Destacan ventas altas en productos de electrónica y mobiliario. Sin embargo, también muestra inventario con baja rotación en juguetes y artículos de entretenimiento.

- Tienda 4: A pesar de que algunos productos tienen ventas decentes, en general muestra menor volumen en comparación con otras sucursales, lo que refuerza lo observado en los ingresos y calificaciones.

---

### Coste de envío promedio



Tienda 1 presenta el costo de envío promedio más alto, lo que puede reducir su competitividad o margen de ganancia.

Tienda 4, en contraste, tiene el costo más bajo, lo cual puede resultar más atractivo para clientes sensibles al precio o para estrategias de envío gratuito.

Tienda 2 y Tienda 3 se ubican en un punto medio, con costes relativamente controlados.

---

### Desempeño geográfico






Tienda 1 y Tienda 3 muestran una distribución más concentrada de puntos rojos, lo que sugiere que poseen zonas específicas de alta rentabilidad.

Tienda 2 presenta una dispersión más homogénea, con puntos mayormente en colores medios, indicando ventas moderadas pero constantes.

Tienda 4 tiene una mayor concentración de puntos azulados, lo que refleja que sus ubicaciones generan ingresos menores en promedio.

## Comparación general y evaluación

### Resumen comparativo por tienda

**Evaluación de fortalezas y debilidades**

- **Tienda 1**: Buen ingreso, pero baja calificación y costo logístico elevado. Debilidad en reputación.
- **Tienda 2**: Perfil equilibrado, pero sin destacar en ningún punto crítico.
- **Tienda 3**: Alto ingreso, buena calificación y desempeño geográfico notable. Perfil fuerte en general.
- **Tienda 4**: Bajo desempeño general, aunque con ventajas en costo de envío. La rentabilidad no compensa las debilidades.

**Síntesis evaluativa**

En conjunto, los datos sugieren que Tienda 3 posee el perfil más sólido y sostenible, mientras que Tienda 4 muestra consistentemente un rendimiento inferior en aspectos clave como ingresos, ventas por producto y satisfacción del cliente. La Tienda 1, aunque genera buenos ingresos, enfrenta riesgos por su baja evaluación y costos elevados.

---

## Conclusión y recomendación

Con base en el análisis integral realizado, se concluye que la **Tienda 3** representa la opción más sólida para mantenerse en operación, mientras que **Tienda 4** se perfila como la mejor candidata para cierre.

**¿Por qué Tienda 4?**

Los datos indican que presenta un rendimiento inferior en varios aspectos clave:

- Menores ingresos totales, lo que impacta directamente la rentabilidad.
- Aunque tiene el costo de envío más bajo, esto no se traduce en mayor volumen de ventas ni en calificaciones destacadas.
- Su portafolio de productos se enfoca en categorías de menor valor, como juguetes y artículos para el hogar, con baja rotación y escasa diferenciación frente a la competencia.
- En términos de desempeño geográfico, sus puntos de venta muestran ingresos bajos en la mayoría de las ubicaciones.

**En contraste, la Tienda 3:**

- Genera altos ingresos y cuenta con la calificación más alta por parte de los clientes, lo cual evidencia una combinación efectiva de ventas y satisfacción.
- Ofrece una línea de productos enfocada en muebles y electrónicos, con alta demanda.
- Tiene un costo logístico moderado y una fuerte presencia en zonas geográficas estratégicas.

**Recomendación final:**

Se sugiere al Sr. Juan cerrar la Tienda 4 y redirigir sus recursos hacia el fortalecimiento de las demás sucursales particularmente la Tienda 3 para maximizar el retorno de inversión y construir una base sólida para su nuevo negocio.
